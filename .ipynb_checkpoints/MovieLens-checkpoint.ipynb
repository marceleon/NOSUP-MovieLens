##### Vemos que pasa con las calificaciones...

```python
#Cargamos las Calificaciones de los usuarios
ratings=pd.read_csv('./data/ratings.csv')
```

```python
print(ratings.shape)
display(ratings.describe())
display(ratings.sample(5))
```

```python
#Grafipo películas por usuario
mvsxusr=ratings.userId.value_counts()
mvsxusr=mvsxusr.reset_index()
mvsxusr.columns=['userId','Cant_Movies']
display(mvsxusr.sample(5))
display(mvsxusr.Cant_Movies.describe())
```

```python
sns.distplot(a=mvsxusr.Cant_Movies, bins=np.arange(0,200,10), kde=False)
```

```python
#Calculo el Rating Promedio por Película
ratings_pp=ratings.pivot_table(index='movieId', values='rating', aggfunc=['count','mean'])
ratings_pp=ratings_pp.reset_index()
ratings_pp.columns=['movieId','rating_cant', 'rating_prm']
ratings_pp.set_index('movieId', inplace=True)

display(ratings_pp.sample(5))
display(ratings_pp.describe())
```

```python
sns.distplot(a=ratings_pp.rating_prm, bins=np.arange(0,6,1), kde=False)
```

[Obs]

* Hay algo menos de 140mil usuarios que rankearon estas películas 20 millones de veces, viendo en promedio 144 pelis. 
* La media de Rating por peli es 3.1 y por visualición 747 veces.
* Por una cuestión de performance vamos a considerar SOLO películas que hayan obetnido puntuación promedio de 4 en adelante. Quedan algo mas de 1700 películas películas vistas 3 millones de veces. 

```python
#Por una cuestión de performance vamos a considerar SOLO películas que hayan sido con puntuación promedio de 4 en adelante.
ratings_apr=pd.merge(left=ratings, right=ratings_pp, how='left', left_on='movieId', right_on='movieId', right_index=True)
print(ratings_apr.shape)
display(ratings_apr.sample(5))
```

```python
ratings_apr=ratings_apr[ (ratings_apr.rating_prm>=4) ]
```

```python
print(ratings_apr.shape)
print('Pelíuclas:' ,ratings_apr.movieId.value_counts().count())
display(ratings_apr.sample(5))
```

[Obs]
#### Vamos a aplicar el algoritmo APRIORI para identificar reglas en este subconjunto de visualizaciones, considerando:
* Nuestros TICKET será cada USUARIO y los ITEMS las películas que rankeó
* Tenemos de 135mil usuarios/ticket de donde extraeremos asociaciones entre películas con los siguientes criterios:

    1. Ambas películas deben haber sido vistas, en conjunto, por al menos en el 0.01% de los usuarios (1300 txts aprox), podría ser menos pero implica mucho tiempo de proceso. (Soporte)
    
    2. La peli recomendada debe haber sido vista por el 50% de los usuarios que vieron la otra peli, al menos. (Confianza)


```python
#Preparo los datos para el algoritmo apriori
ratings_apr.sort_values(by='userId', axis=0, ascending=True, inplace=True)
display(ratings_apr.head(5))
display(ratings_apr.tail(5))
```

```python
transactions=[]
for usr, mvs in groupby(ratings_apr[ ['userId','movieId'] ].values, lambda x: x[0]):
    transactions.append([mm[1] for mm in mvs])

print(transactions[1])
len(transactions)
```

```python
itemsets, rules = apriori(transactions, min_support=0.01,  min_confidence=0.5, max_length=2)
```

```python
#Imprimo algunas reglas
rules=sorted(rules, key=lambda rule: rule.confidence)
for rule in rules:
    if rule.lift >=20:
        print(rule)
```

```python
print('Reglas:',len(rules))
```

```python
#Cargamos las Reglas en un DataFrame y anexamos datos de las Pelis
Reglas=pd.DataFrame(columns=['lhs','rhs','supp','conf','lift','conv'],
                    data=[[r.lhs[0], r.rhs[0], r.support, r.confidence, r.lift, r.conviction] for r in rules])
```

```python
print(Reglas.shape)
Reglas.head(5)
```

```python
#Anexamos datos de las Películas
Reglas=pd.merge(left=Reglas, right=movies[ ['movieId','title','Generos'] ], 
                how='left', left_on='lhs', right_on='movieId')
Reglas=pd.merge(left=Reglas, right=ratings_pp, 
                how='left', left_on='lhs', right_on='movieId')
Reglas=pd.merge(left=Reglas, right=movies[ ['movieId','title','Generos'] ], 
                how='left', left_on='rhs', right_on='movieId')
Reglas=pd.merge(left=Reglas, right=ratings_pp, 
                how='left', left_on='rhs', right_on='movieId')
```

```python
Reglas.shape
```

```python
Reglas.head(5)
```

```python
#Analizo Precedente
Reglas_Prcdnte=Reglas.lhs.value_counts().to_frame()
Reglas_Prcdnte.reset_index(inplace=True)
Reglas_Prcdnte.columns=['movieId','cant']
```

```python
print(Reglas_Prcdnte.shape)
print(Reglas_Prcdnte.cant.mean())
pd.merge(left=Reglas_Prcdnte.sort_values(by='cant',ascending=False).head(5), 
         right=movies[['movieId','title']], how='left', left_on='movieId', right_on='movieId')
```

```python
#Analizo Consecuente
Reglas_Cnscte=Reglas.rhs.value_counts().to_frame()
Reglas_Cnscte.reset_index(inplace=True)
Reglas_Cnscte.columns=['movieId','cant']
```

```python
print(Reglas_Cnscte.shape)
print(Reglas_Cnscte.cant.mean())
pd.merge(left=Reglas_Cnscte.sort_values(by='cant',ascending=False).head(5), 
         right=movies[['movieId','title']], how='left', left_on='movieId', right_on='movieId')
```
